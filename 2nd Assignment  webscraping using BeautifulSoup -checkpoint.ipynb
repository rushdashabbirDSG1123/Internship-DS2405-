# 1) Write a python program to display IMDB’s Top rated 100 Indian movies’ data
https://www.imdb.com/list/ls056092300/ (i.e. name, rating, year ofrelease) and make data frame.


```python
import pandas as pd
import requests
import bs4
from bs4 import BeautifulSoup
import regex as re
```

```python
url='https://www.imdb.com/list/ls056092300/'
headers={'User-Agent':'Mozilla/5.0'}
# send request to the web page
web=requests.get(url,headers=headers)
print(web)

#collect content from web page
soup=BeautifulSoup(web.content,'html.parser')

# extract movie nam
movie_name=soup.find('div',class_="ipc-title ipc-title--base ipc-title--title ipc-title-link-no-icon ipc-title--on-textPrimary sc-b189961a-9 iALATN dli-title")
print(movie_name.text)

# extract Rating
rating=soup.find('div',class_="sc-e2dbc1a3-0 ajrIH sc-b189961a-2 fkPBP dli-ratings-container")
print(rating.text.replace("\xa0",""))

#extract Releasing Year
year=soup.find('div',class_="sc-b189961a-7 feoqjK dli-title-metadata")
Relese_year= year.text
pattern=r'(\d{4}|\d{2}\w{1})'
year_release=re.sub(pattern,r"\1 ",Relese_year)
print(year_release)


# create movie name list
movie_names = []
for mov in soup.find_all('div',class_="ipc-title ipc-title--base ipc-title--title ipc-title-link-no-icon ipc-title--on-textPrimary sc-b189961a-9 iALATN dli-title"):
    movie_names.append(mov.text)
#print(movie_names)


# create Movie rating list
movie_ratings = []
for rat in soup.find_all('div',class_="sc-e2dbc1a3-0 ajrIH sc-b189961a-2 fkPBP dli-ratings-container"):
    movie_ratings.append(rat.text.replace("\xa0",''))
#print(movie_ratings)


# create movire Releasing Year
movie_years = []
for yer in soup.find_all('div', class_="sc-b189961a-7 feoqjK dli-title-metadata"):
    movie_years.append(yer.text)

years_with_spaces = []
for year in movie_years:
    years_with_spaces.append(re.sub(r'(\d{4}|\d{2}\w{1})', r'\1 ', year))
#print(years_with_spaces)

# create Data Frame 
df=pd.DataFrame({'Movie Name':movie_names,'Movei Rating':movie_ratings,'Relase year':years_with_spaces})
print(df)
```

```python
def scrape_ten_product_details():
    url='https://www.bewakoof.com/bestseller?sort=popular%20'
    response_web=requests.get(url)
    response_web
    soup=BeautifulSoup(response_web.content,'html.parser')
    #print(soup)
    
    #scrape Product Details
    
    Product_name=[nam.text for nam in soup.find_all('div', class_="productNaming bkf-ellipsis")[:10]]
    Price=[pri.text for pri in soup.find_all('div',class_="loyaltyPriceBox")[:10]]
    image_url=[img['src'] for img in soup.find_all('img',class_="productImgTag")[:10]]
    
    #dataFrame
    
    df_bewakoof=pd.DataFrame({'Product name':Product_name,'Product Price':Price,'Product Images':image_url})
    return df_bewakoof

scrape_ten_product_details()
```
