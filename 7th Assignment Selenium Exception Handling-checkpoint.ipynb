# Question 2

2. In the above question, now scrape the following details of each product listed in first 3 pages of your search
results and save it in a data frame and csv. In case if any product has less than 3 pages in search results then
scrape all the products available under that product name. Details to be scraped are: "Brand
Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and
“Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“. 

```python
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
import time
import pandas as pd

Product_url = []
start = 0
end = 3

# Scrape pages
for page in range(start, end):
    time.sleep(5)
    urls = driver.find_elements(By.XPATH, '//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for url in urls:
        Product_url.append(url.get_attribute('href'))
    
    time.sleep(3)
    try:
        next_button = driver.find_element(By.XPATH, '/html/body/div[1]/div[1]/div[1]/div[1]/div/span[1]/div[1]/div[67]/div/div/span/a[3]')
        next_button.click()
    except NoSuchElementException as e:
        print("Exception raised:", e)
        break

print(len(Product_url))
#print(Product_url)

brand_nam = []
Name_product = []
Price_product = []
return_detail = []
Expected_delivery = []
Availability = []

for i in Product_url:
    driver.get(i)
    
    # Scrape Brand name
    time.sleep(5)
    try:
        brand = driver.find_element(By.XPATH, "//a[@id='bylineInfo']")
        brand_nam.append(brand.text)
    except NoSuchElementException as e:
        brand_nam.append('Not Present or "-"')
    
    
    # Product name
    time.sleep(6)
    try:
        product = driver.find_element(By.XPATH, '//span[@id="productTitle"]')
        Name_product.append(product.text)
    except NoSuchElementException as e:
        Name_product.append("-")
    
    # Scrape Product price
    time.sleep(5)
    try:
        price = driver.find_element(By.XPATH, '//span[@class="a-price aok-align-center reinventPricePriceToPayMargin priceToPay"]')
        Price_product.append(price.text)
    except NoSuchElementException as e:
        Price_product.append("-")
    
    # Scrape Return policy
    time.sleep(5)
    try:
        retrn = driver.find_element(By.XPATH, '//div[@id="RETURNS_POLICY"]')
        return_detail.append(retrn.text)
    except NoSuchElementException as e:
        return_detail.append("-")
    
    # Scrape Expected delivery
    time.sleep(5)
    try:
        delivery = driver.find_element(By.XPATH, '//span[@class="a-text-bold"]')
        Expected_delivery.append(delivery.text)
    except NoSuchElementException as e:
        Expected_delivery.append("-")
    
    # Scrape Availability
    time.sleep(5)
    try:
        available = driver.find_element(By.XPATH, '//span[@class="a-size-medium a-color-success"]')
        Availability.append(available.text)
    except NoSuchElementException as e:
        Availability.append("-")

print(len(brand_nam))
print(len(Name_product))
print(len(Price_product))
print(len(return_detail))
print(len(Expected_delivery))
print(len(Availability))

# Create DataFrame
Product_df = pd.DataFrame({
    "Brand Name": brand_nam,
    "Product Name": Name_product,
    "Product Price": Price_product,
    "Return/Exchange Policy": return_detail,
    "Expected Delivery": Expected_delivery,
    "Availability": Availability
})

# Print the DataFrame
print(Product_df)
```

```python
import time
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
#exceptions
from selenium.common.exceptions import NoSuchElementException ,TimeoutException,StaleElementReferenceException
import pandas as pd
import numpy as np
import regex as re

# run browser
driver=webdriver.Chrome() 
time.sleep(2)

# opened google map
driver.get("https://www.google.co.in/maps")
time.sleep(3)

# find search bar
search = driver.find_element(By.XPATH,'//input[@class="fontBodyMedium searchboxinput xiQnY "]')                       
search.clear() 

#search city
time.sleep(5)
search.send_keys("Pune")                                                     
search_button = driver.find_element(By.XPATH,'//button[@class="mL3xi"]')               
search_button.click()                                                             


#scrape longitude and latitude with the help of regex module
time.sleep(5)

try:
    url_lat_long = driver.current_url
    print("URL Extracted: ", url_lat_long)
    
    lat_long = re.findall(r'@(.*)data',url_lat_long)
    if len(lat_long):  # if here 2 values
        lat_long_list = lat_long[0].split(",")
        
        if len(lat_long_list)>=2:
            lat = lat_long_list[0]
            long = lat_long_list[1]
        print("Latitude = {}, Longitude = {}".format(lat, long))

except Exception as e:
        print("Error: ",e)

driver.quit()
```

```python
Laptop_Name = []
Operating_sys = []
Display = []
Resolution=[]
Processor = []

#for scrape the laptop name
time.sleep(3)
try:
    lap_tags=driver.find_elements(By.XPATH,'//div[@class="rh_gr_top_middle mb10 colored_rate_bar"]')
    for l in lap_tags:
        Laptop_Name.append(l.text)
except NoSuchElementException as e:
    Laptop_Name.append("-")
    
#for scrape the Operating system
time.sleep(3)
try:
    os=driver.find_elements(By.XPATH,'//span[@class="meta_v_value"]')
    for o in os:
        Operating_sys.append(o.text.strip())
except NoSuchElementException as e:
    Operating_sys.append("-")
    
    
for o in Operating_sys[:]:
    if 'Windows' not in o:
        Operating_sys.remove(o)
        
#for scrape the laptop Display
time.sleep(3)
try:
    dis=driver.find_elements(By.XPATH,'//span[@class="meta_v_value"]')
    for d in dis:
        Display.append(d.text)
except NoSuchElementException as e:
    Display.append("-")
    
for d in Display[:]:
    if '.' not in d:
        Display.remove(d)

        
#for scrape the laptop Reslution
time.sleep(3)
try:
    reso=driver.find_elements(By.XPATH,'//span[@class="meta_v_value"]')
    for r in reso:
        Resolution.append(r.text)
except NoSuchElementException as e:
    Resolution.append("-")
    
for r in Resolution[:]:
    if 'x' not in r:
        Resolution.remove(r)
        
#for scrape the laptop Processor
time.sleep(3)
try:
    pro=driver.find_elements(By.XPATH,'//span[@class="meta_v_value"]')
    for p in pro:
        Processor.append(p.text)
except NoSuchElementException as e:
    Processor.append("-")
new_Processor = []
for P in Processor:
    if 'Ryzen' in P or 'Core' in P or 'Intel' in P or 'AMD' in P:
        new_Processor.append(P.split(':')[0].strip())
new_Processor = [P.split(':')[0].strip() for P in Processor if P.startswith('AMD Ryzen')]


print(len(Laptop_Name))
print(len(Operating_sys))
print(len(Display))
print(len(Resolution))
print(len(new_Processor))
```

```python
best_gaming_laptop_df=pd.DataFrame({
    "Laptop Name":Laptop_Name[:3],
    "Operating system":Operating_sys[:3],
    "Display":Display[:3],
    "Resolution":Resolution[:3],
    "Laptop Precessor":new_Processor[:3]
})
best_gaming_laptop_df
```

```python
import time
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import warnings
warnings.filterwarnings('ignore')

#exceptions
from selenium.common.exceptions import NoSuchElementException ,TimeoutException
import pandas as pd
import numpy as np

driver=webdriver.Chrome()
driver.maximize_window()
driver.get('https://www.youtube.com')

# Finding search menu by xpath
Search=driver.find_element(By.XPATH,'//input[@id="search"]')
# Feeding input video name by user to search menu through send keys
Search.send_keys('motivated english song')

# Finding Search button for clicking through xpath
Search_button=driver.find_element(By.XPATH,'//button[@id="search-icon-legacy"]/yt-icon')  
# Clicking search button
Search_button.click()
time.sleep(2)


#click on vedio
link = driver.find_element(By.XPATH,'//a[@title="Sia - Unstoppable (Lyrics)"]')
link.click()
time.sleep(3)


#scroll for extracting more Comments
for _ in range(1500):
    driver.execute_script("window.scrollBy(0,1500)")
    
# Empty lists to scrap data
Comments = []
Comment_posted_time = []
Likes = []


#extracting comment on video
time.sleep(5)
comment = driver.find_elements(By.ID,"content-text")
for com in comment:
    if com.text is None:
        Comments.append('--')
    else:
        Comments.append(com.text)
        
#extracting comment times
time.sleep(5)
c_time = driver.find_elements(By.ID,"published-time-text")
for t in c_time:
    Comment_posted_time.append(t.text)
    
#extracting comment likes
time.sleep(5)
c_likes= driver.find_elements(By.XPATH,'//span[@class="style-scope ytd-comment-engagement-bar"]')
for l in c_likes:
    Likes.append(l.text)
    

print(len(Comments))
print(len(Comment_posted_time))
print(len(Likes))


# making Data collecting the dictionary
Comments_df=({
    "Vedio Comments":Comments,
    "Comments Timing":Comment_posted_time,
    "Comments Likes":Likes})


# making Data Frame
Comments_df = pd.DataFrame(Comments_df)
# we want to scrape only 500 
Comments_df = Comments_df.head(500)
Comments_df
```

9. Write a python program to scrape a data for all available Hostels from https://www.hostelworld.com/ in
“London” location. You have to scrape hostel name, distance from city centre, ratings, total reviews, overall
reviews, privates from price, dorms from price, facilities and property description. 


```python
import time
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import warnings
warnings.filterwarnings('ignore')

#exceptions
from selenium.common.exceptions import NoSuchElementException ,TimeoutException,StaleElementReferenceException
import pandas as pd
import numpy as np


# Connecting to webdriver
driver=webdriver.Chrome()
time.sleep(1)
driver.maximize_window()

# Opening url in chrome browser
url="https://www.hostelworld.com"
driver.get(url)
time.sleep(2)

#locating the location search bar
Search = driver.find_element(By.XPATH,'//input[@class="native-input"]')

# Sending input London in search bar
Search.send_keys("London")
time.sleep(1)

#select london
time.sleep(2)
try:
    london = driver.find_element(By.XPATH,'//button[@class="item-content"]').click()
except StaleElementReferenceException as e:
    print("Exception Raised")

# click on search button
Search_button = driver.find_element(By.XPATH,'//button[@class="large-button btn-content"]')
Search_button.click()
```

```python
# Make empty lists
name = []
distance = []
rating = []
review = []
overview_review = []
private_price  = []
dorm_price = []
facility=[]
discription=[]
hotel_url=[]

name_tag= driver.find_elements(By.XPATH,'//div[@class="property-name"]/span')
dist_tag= driver.find_elements(By.XPATH,'//div[@class="property-distance"]/span[2]')
rate_tag= driver.find_elements(By.XPATH,'//span[@class="number"]')
rev_tag= driver.find_elements(By.XPATH,'//span[@class="keyword"]')
ovrrev_tag= driver.find_elements(By.XPATH,'//div[@class="review"]/span')
pvt_tag= driver.find_elements(By.XPATH,'//div[@class="accommodation-price single-row"]/strong') 
dor_tag= driver.find_elements(By.XPATH,'//strong[@class="current"]')

min_length = min(len(name_tag), len(dist_tag), len(rate_tag), len(rev_tag), len(ovrrev_tag), len(pvt_tag), len(dor_tag))


for i in range(min_length):
    name.append(name_tag[i].text)
    distance.append(dist_tag[i].text)
    rating.append(rate_tag[i].text)
    review.append(rev_tag[i].text)
    overview_review.append(ovrrev_tag[i].text)
    private_price.append(pvt_tag[i].text)
    dorm_price.append(dor_tag[i].text)
    
# hotel URL
for page in range(0,1):
    urls=driver.find_elements(By.XPATH,'//a[@class="property-card-container horizontal"]')
    for url in urls:
        hotel_url.append(url.get_attribute('href'))
```

```python
#click on hotel URL for scrape description
try:
    click_hotel_url=driver.find_element(By.XPATH,'//a[@class="property-card-container horizontal"]').click()
except NoSuchElementException as e:
    print("Exception Raised")
```

```python
#scrape Facility DATA

for f in hotel_url:
    driver.get(f)
    time.sleep(3)
    try:
        fac=driver.find_element(By.XPATH,'//div[@class="facilities-category-container"]/div[2]/span')
        facility.append(fac.text)
    except NoSuchElementException as e:
        facility.append("No element")
        
        
#scrape discription
for d in hotel_url:
    driver.get(d)
    time.sleep(3)
    try:
        des=driver.find_element(By.XPATH,'//p[@class="hw-says-description"]')
        discription.append(des.text)
    except NoSuchElementException as e:
        discription.append("No element")
        
#check length of the data
print(len(name))
print(len(distance))
print(len(rating))
print(len(review))
print(len(overview_review))
print(len(private_price))
print(len(dorm_price))
print(len(hotel_url))
print(len(facility))
print(len(discription))

driver.quit()
```
