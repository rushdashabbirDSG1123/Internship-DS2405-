```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
import time
import pandas as pd

driver = webdriver.Chrome()
driver.maximize_window()
driver.get('https://www.flipkart.com/')

# find search input sneakers
search_input = driver.find_elements(By.CLASS_NAME, "Pke_EE")
search_input[0].send_keys('sneakers')

# click search button
search_button = driver.find_elements(By.CLASS_NAME, '_2iLD__')
for p in search_button:
    p.click()

# create list
time.sleep(6)
product_brand = []
product_description = []
product_price = []

# scrape data
start = 1
end = 10
for web in range(start, end):
    brand = driver.find_elements(By.XPATH, '//div[@class="syl9yP"]')
    desc = driver.find_elements(By.XPATH, '//a[@class="WKTcLC BwBZTg"]')
    price = driver.find_elements(By.XPATH, '//div[@class="Nx9bqj"]')

    # Find the minimum length among the three lists
    min_len = min(len(brand), len(desc), len(price))

    for i in range(min_len):
        product_brand.append(brand[i].text)
        product_description.append(desc[i].text)
        product_price.append(price[i].text)

# make Data Frame
sneaker_df = pd.DataFrame({
    "Brand name": product_brand,
    "Sneaker Description": product_description,
    "Price": product_price
})

print(sneaker_df)
driver.quit()
sneaker_df[:100]
```

The above task will be done in following steps:

```python
import time
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
import pandas as pd
import numpy as np

driver=webdriver.Chrome()
driver.maximize_window()
                                        # STEP 1
driver.get('https://www.azquotes.com')
                                        # STEP 2
#click on TopQuote
topquote=driver.find_elements(By.XPATH,'/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a')
for click in topquote:
    click.click()
                                        # STEP 3
#scrape data
author=[]
quote=[]
type_quote=[]
start=1
end=11

for page in range(start,end):
    auth=driver.find_elements(By.XPATH,'//div[@class="author"]')
    quo=driver.find_elements(By.XPATH,'//a[@class="title"]')
    typ=driver.find_elements(By.XPATH,'//div[@class="tags"]')
    
    for i in range(len(auth)):
        author.append(auth[i].text)
        quote.append(quo[i].text)
        type_quote.append(typ[i].text)
#make datafrme
author_df=pd.DataFrame({
    "Author Name":author,
    "Quotes":quote,
    "Types Of Quote":type_quote
})
driver.quit()
author_df
```

Q7: Write a python program to display list of respected former Prime Ministers of India (i.e. Name,
Born-Dead, Term of office, Remarks) from https://www.jagranjosh.com/general-knowledge/list-ofall-prime-ministers-of-india-1473165149-1

```python
import time
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
import pandas as pd
import numpy as np

import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By

# open Chrome browser
driver = webdriver.Chrome()

# get web page
driver.get('https://www.jagranjosh.com/general-knowledge/list-of-all-prime-ministers-of-india-1473165149-1')

# search table body elements
tbodies = driver.find_elements(By.XPATH, '//table/tbody')

data = []

for tbody in tbodies:
    trs = tbody.find_elements(By.TAG_NAME, 'tr')
    for tr in trs:
        row = [item.text for item in tr.find_elements(By.XPATH, './/td')]
        if row:
            data.append(row)

# Create a list of columns
column_names = [f"Column {i+1}" for i in range(len(data[0]))]

# Convert the data into DataFrame
priminister_df = pd.DataFrame(data, columns=column_names)

# remove column 1 because its unnecessary
priminister_df = priminister_df.drop(columns=[column_names[0]])

# give the columns name 
priminister_df.columns = ['Prime Minister', 'Born-Dead', 'Term of office','Remark']


# Close chrome driver
driver.quit()
priminister_df
```

```python
import time
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
import pandas as pd
import numpy as np

driver=webdriver.Chrome()
driver.maximize_window()
                                                    #STEP 1
#get web page
driver.get('https://www.motor1.com')

                                                    #STEP 2
#find search bar write 50 most expensive cars
search=driver.find_elements(By.XPATH,'/html/body/div[9]/div[2]/div/div/div[3]/div/div/div/form/input')
search[0].send_keys('50 most expensive cars')
                                                    
# click at search button
search_button=driver.find_elements(By.XPATH,'//button[@class="m1-search-panel-button m1-search-form-button-animate icon-search-svg"]')
for clicks in search_button:
    clicks.click()
    
                                                    #STEP 3
# click on 50 most expensive cars
click_exp=driver.find_elements(By.XPATH,'/html/body/div[9]/div[9]/div/div[1]/div/div/div[1]/div/div[1]')
for c in click_exp:
    c.click()
                                                    #STEP 4
 #crape Car Data
car_name=[]
price=[]
cars=driver.find_elements(By.XPATH,'//h3[@class="subheader"]')
pri=driver.find_elements(By.XPATH,'//strong')

for c,p in zip(cars,pri):
    car_name.append(c.text)
    price.append(p.text)
    
print(len(car_name),len(price))
driver.quit()
# make Data frame
car_df=pd.DataFrame({'Car Name':car_name[:50],
                    "Car prices":price[:50]})

car_df       
```
