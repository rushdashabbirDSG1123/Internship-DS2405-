1) Write a python program to display all the header tags from wikipedia.org

```python
header=soup.find('div',class_="central-textlogo")
header.text.replace("\n",'')
```

```python
header_tag=[]
for i in soup.find_all(['h1', 'h2', 'h3', 'h4', 'h5', 'h6']):
    header_tag.append(i.text.replace("\n",""))
header_tag
```

2) Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name,rating, year
of release) and make data frame.


```python
url = "https://m.imdb.com/chart/top"
headers = {'User-Agent': 'Mozilla/5.0'}
response = requests.get(url, headers=headers)

print(response)
```

```python
soup=BeautifulSoup(response.content,'html.parser')
soup
```

```python
name=soup.find('a',class_="ipc-title-link-wrapper")
print(name.text)
rate=soup.find('span',class_="ipc-rating-star ipc-rating-star--base ipc-rating-star--imdb ratingGroup--imdb-rating")
print(rate.text.replace("\xa0",""))
year=soup.find('span',class_="sc-b189961a-8 kLaxqf cli-title-metadata-item")
print(year.text)
```

```python
movie=[]
for mov in soup.find_all('a',class_="ipc-title-link-wrapper"):
    movie.append(mov.text)
movie
```

```python
Rating=[]
for rate in soup.find_all('span',class_="ipc-rating-star ipc-rating-star--base ipc-rating-star--imdb ratingGroup--imdb-rating"):
    Rating.append(rate.text.replace("\xa0",""))
Rating
```

```python
Releasing_year=[]
for release in soup.find_all('div',class_="sc-b189961a-7 feoqjK cli-title-metadata"):
    Releasing_year.append(release.text)

pattern=r'(\d{4}|\d{2}\w{1})'

Release_date=[re.sub(pattern,r'\1 ',i)for i in Releasing_year]

print(Release_date)
     
```

```python
print(len(movie),len(Rating),len(Release_date))
```

```python
df=pd.DataFrame({'Movie Name':movie,'Movie rating':Rating,'Releasing Year':Release_date})
df
```

3.Write a python program to scrape mentioned details from dineout.co.in : i) Restaurant name
ii) Cuisine iii) Location iv) Ratings v) Image URL.


```python
web='https://www.dineout.co.in/delhi-restaurants/welcome-back'
restore=requests.get(web)
print(restore)
soup=BeautifulSoup(restore.content)
```

```python
restau_name=soup.find('a',class_="restnt-name ellipsis")
print(restau_name.text)
cuisin=soup.find('span',class_="double-line-ellipsis")
print(cuisin.text)
location=soup.find('div',class_="restnt-loc ellipsis")
print(location.text)
rating=soup.find('div',class_="restnt-rating rating-4")
print(rating.text)
image=soup.find('img',class_="no-img")
print(image)
```

```python
restau_name=[]
for rest in soup.find_all('a',class_="restnt-name ellipsis"):
    restau_name.append(rest.text)
restau_name
```

```python
cuisin=[]
for cuis in soup.find_all('span',class_="double-line-ellipsis"):
    cuisin.append(cuis.text)
cuisin
```

```python
location=[]
for loc in soup.find_all('div',class_="restnt-loc ellipsis"):
    location.append(loc.text)
location
```

```python
rating=[]
for rat in soup.find_all('div',class_="restnt-rating rating-4"):
    rating.append(rat.text)
rating
```

```python
image=[]
for img in soup.find_all('img',class_="no-img"):
    image.append(img['data-src'])
image
```

```python
df1=pd.DataFrame({'Restaurent Name':restau_name,'Cuisin':cuisin,'Location':location,'Images Of Restaurent':image})
df1
```

4) Write s python program to display list ofrespected former presidents of India (i.e.
Name , Term of office) fromhttps://presidentofindia.nic.in/former-presidents
andmake data frame.

```python
url='https://presidentofindia.nic.in/former-presidents'
```

```python
content=requests.get(url)
content
```

```python
soup=BeautifulSoup(content.content)
soup
```

```python
Name_president=[]
for nam in soup.find_all('div',class_="desc-sec"):
    Name_president.append(nam.text.replace('\n',''))
Name_president
```

```python
df2=pd.DataFrame({'Name_president':Name_president})
df2
```
